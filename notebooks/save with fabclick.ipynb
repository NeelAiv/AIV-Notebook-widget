```
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(10)  # Example data
plt.plot(range(10), data)
plt.title('Random Line Graph')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```
