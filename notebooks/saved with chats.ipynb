```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = [1, 2, 3]
y = [2, 4, 6]
z = [2, 4, 6]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.bar3d(x, y, [0 for _ in range(len(x))], [1 for _ in range(len(x))], [z[i] - y[i] for i in range(len(x))], color='b')

plt.show()
```
