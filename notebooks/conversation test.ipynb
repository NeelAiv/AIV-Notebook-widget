```
import matplotlib.pyplot as plt

data = [10, 15, 7, 12, 20]
labels = ['A', 'B', 'C', 'D', 'E']

plt.bar(labels, data)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Graph')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('3D Line Graph')
plt.show()
```
